```python
import re
stringData = "f sdf jsj flslkf sdhf sjkfh yes kjsdh fjhsd"
pattern = r"yes"
isPatternFound = re.search(pattern,stringData)
print(type(isPatternFound))
if isPatternFound:
    print('Match found')
else:
    print('Match not found')
```

```python
patternData = "\d\d"
print(patternData)

pattern = r"\d\d\d\d\d\d\d\d\d\d"  # Matches any ten consecutive digits
print(pattern)
text = "My Phone number is 1234567890"
match = re.search(pattern, text)
print(match)
if match:
    print('Match found',match.group())
else:
    print('Match not found')
```

```python
pattern = r"\W"  # Matches any non-word character
text = "Hello, world!"
matches = re.findall(pattern, text)
print(matches)
print("Matches:", matches)
```

```python
pattern = r"\W"  # Matches any non-word character
text = "Hello, world!"
print(text[0:5])
matches = re.findall(pattern, text)
print(matches)
print("Matches:", len(matches))
```
