```python
from colorama import Fore, Style

person = {"Name": "Alice", "Age": 30, "Occupation": "Engineer","Name": "Alice1", "Age": 31, "Occupation": "Engineer1"}

for key, value in person.items():
    print(f"{Fore.BLUE}{key}{Style.RESET_ALL}: {value}")
```
