```python
givenstring="sathik sathik sathik Lorem ipsum dolor! diam amet, consetetur Lorem magna. sed diam nonumy eirmod tempor. diam et labore? et diam magna. et diam amet."
givenstring
```

```python
class TextAnalyzer(object):
    
    def __init__(self,textData):
        formattext = textData.replace("!","").replace(".","").replace("?","").replace("'","").replace("\"","")
        formattext = formattext.lower()
        self.textData = formattext

    def freqAll(self): 
        dictFreq = {}
        listData = self.textData.split(" ")
        for data in set(listData):
            dictFreq[data]=listData.count(data)
        return dictFreq

    def freqword(self,word):
        freqDict = self.freqAll()
        if word in freqDict:
            return freqDict[word]
        else:
            return 0
        
```

```python
textData = TextAnalyzer(givenstring)
textData.freqAll()
textData.freqword('sathik')
```
