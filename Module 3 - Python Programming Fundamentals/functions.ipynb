```python
def printStatement(statements):
    for i,s in enumerate(statements):
        print(f"---{i}--->{s}")

alb_rating = [10.25,8.5,27.1]
printStatement(alb_rating)
```

```python
def printName(*names):
    for n in names:
        print(f"---{n}--->")
printName("aa","bb","cc")
```
